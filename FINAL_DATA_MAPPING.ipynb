```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
map1=pd.read_csv('map.csv')
```

```python
position=pd.read_csv("position.csv")
```

```python
position
```

```python
map1
```

```python
position.lat=position.lat/100
position.log=position.log/100
```

```python
position.columns
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(position[['lat','log']],
                                                     position['Sun_Elevation'], test_size=0.33, random_state=42)
model1=XGBRegressor()
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

```python
X_train, X_test, y_train, y_test = train_test_split(position[['lat','log']],
                                                     position['Sat_Elevation'], test_size=0.33, random_state=42)
model2=XGBRegressor()
model2.fit(X_train,y_train)
model2.score(X_test,y_test)
```

```python
X_train, X_test, y_train, y_test = train_test_split(position[['lat','log']],
                                                     position['Sat_Azimuth'], test_size=0.33, random_state=42)
model3=XGBRegressor()
model3.fit(X_train,y_train)
model3.score(X_test,y_test)
```

```python
X_train, X_test, y_train, y_test = train_test_split(position[['lat','log']],
                                                     position['Sun_Azimuth'], test_size=0.33, random_state=42)
model4=XGBRegressor()
model4.fit(X_train,y_train)
model4.score(X_test,y_test)
```

```python
sun_elevation=model1.predict(map1[['lat','log']])
sat_elevation=model2.predict(map1[['lat','log']])
sat_azimuth=model3.predict(map1[['lat','log']])
sun_azimuth=model4.predict(map1[['lat','log']])
```

```python
map1['Sun_Elevation']=sun_elevation
map1['Sat_Elevation']=sat_elevation
map1['Sat_Azimuth']=sat_azimuth
map1['Sun_Azimuth']=sun_azimuth
```

```python
map1
```

```python
map1.to_csv('train.csv')
```
