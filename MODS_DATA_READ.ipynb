```python
log=log.values.reshape(-1,1)
lat=lat.values.reshape(-1,1)
cot=cot.values.reshape(-1,1)
cer=cer.values.reshape(-1,1)
```

```python
cot
```

```python
log=pd.DataFrame(log)
lat=pd.DataFrame(lat)
cer=pd.DataFrame(cer)
cot=pd.DataFrame(cot)
```

```python
df=pd.concat([log,lat,cer,cot],axis=1)
```

```python
df.columns=["log","lat","cot","cer"]
```

```python
df.shape
```

```python
df=df[df.cot>0]
```

```python
df=df[df.cer>0]
```

```python
df.to_csv("data.csv")
```
